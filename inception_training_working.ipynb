```python
def load_data(data_path: str, filenames: list):
    
    os.chdir(data_path)
    l = []
    for filename in filenames:
        if (os.path.isfile(filename)):
            arr = np.load(filename)
            l.append(arr)
        else:
            print(filename)
    return np.stack(l)
```

### Load Positive Train / Test / Val Data

```python
def load_neg_dataframe(data_path: str):
    train = []
    test = []
    val = []
    df = pd.read_csv(data_path,sep=',',header=None,encoding = "utf-8-sig")
    for index,row in df.iterrows():
#         print(row[0],row[1],row[2])
        if row[0][0] != 'N':
            continue
        missing_list = ['N330','N140','N309','N245','N63','N204','N149','N617','N211','N105','N267','N437','N295','N494', \
                        'N196','N218','N116','N53','N283','N13','N277','N180','N58','N625','N276','N259','N132']
        
        
            
        if (not pd.isnull(row[0])):
            if row[0] in missing_list:
                continue
            train_ID = row[0]+'.npy'
            train.append(train_ID)

            
        if (not pd.isnull(row[1])):
            if row[1] in missing_list:
                continue
            test_ID = row[1]+'.npy'
            test.append(test_ID)

            
        if (not pd.isnull(row[2])):
            if row[2] in missing_list:
                continue
            val_ID = row[2]+'.npy'
            val.append(val_ID)


    print('train_data: '+str(len(train)))
    print('test_data: '+str(len(test)))
    print('val_data: '+str(len(val)))
    return train,test,val
```

```python
def load_pos_dataframe(data_path: str):
    train = []
    test = []
    val = []
    df = pd.read_csv(data_path,sep=',',header=None,encoding = "utf-8-sig")
    for index,row in df.iterrows():
        if row[0][0] != 'P':
            continue
#         if row[0] == 'P115': #remove non-existent entries
#             continue
#         if row[0] == 'P266': #remove non-existent entries
#             continue
#         if row[0] == 'P1': #remove non-existent entries
#             continue
#         if row[0] == 'P158': #remove non-existent entries
#             continue
#         if row[0] == 'P58': #remove non-existent entries
#             continue
        ID = row[0]+'.npy'
        missing_list = ['P115','P266','P1','P158','P58','P210','P77','P296','P225','P182','P248','P131','P93',\
                        'P203','P95','P286','P299','P74','P211','P177','P205','P19','P256','P157','P91','P276',\
                       'P62','P135','P300','P173','P193','P14','P48','P36','P166','P101','P290','P201','P242']
        if row[0] in missing_list:
            continue
        data_type = row[3]

        if (data_type == 'Train'):
            train.append(ID)
        if (data_type == 'Test'):
            test.append(ID)
        if (data_type == 'Val'):
            val.append(ID)

    print('train_data: '+str(len(train)))
    print('test_data: '+str(len(test)))
    print('val_data: '+str(len(val)))
    return train,test,val
```
